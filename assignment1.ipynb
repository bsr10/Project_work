```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Task 1: Data Preparation

```python
#Reading the file
sw = pd.read_csv("StarWars.csv", encoding= 'unicode_escape', header=[0,1], error_bad_lines = False, index_col=0) 
sw.head()                      
```

```python
#Renaming columns for accessibility
sw = pd.read_csv("StarWars.csv", encoding= 'unicode_escape', header=[1], error_bad_lines = False)  
sw.rename(columns={ sw.columns[0]: "RespondentID", 
    sw.columns[1]: "Have you seen any of the 6 films in the Star Wars franchise?",
    sw.columns[2]: "Do you consider yourself to be a fan of the Star Wars film franchise?",
    sw.columns[3]: "seen_1",
    sw.columns[4]: "seen_2",
    sw.columns[5]: "seen_3",
    sw.columns[6]: "seen_4",
    sw.columns[7]: "seen_5",
    sw.columns[8]: "seen_6",
    sw.columns[9]: "rank_1",
    sw.columns[10]: "rank_2",
    sw.columns[11]: "rank_3",
    sw.columns[12]: "rank_4",
    sw.columns[13]: "rank_5",
    sw.columns[14]: "rank_6",
    sw.columns[29]: "Which character shot first?",
    sw.columns[30]: "Are you familiar with the Expanded Universe?",
    sw.columns[31]: "Do you consider yourself to be a fan of the Expanded Universe?",
    sw.columns[32]: "Do you consider yourself to be a fan of the Star Trek franchise?",
    sw.columns[33]: "Gender",
    sw.columns[34]: "Age",
    sw.columns[35]: "Household Income",
    sw.columns[36]: "Education",
    sw.columns[37]: "Location (Census Region)",
    }, inplace= True)

#[seen_1] means 'Have you seen the first movie?'
#[rank_1] means 'Which position would you rank the first movie out of all 6?'
#[Han Solo] is the ranking of the character by the respondent
```

```python
#Checking datatype of each variable
sw.dtypes
```

```python
#Checking dimensions of data set
sw.shape
```

```python
#Viewing (unique) value counts for each column
for i in sw.columns:
    x = sw[i].value_counts(dropna=False)
    print("Column name is:",i,"and its value is:",x)
    print()
    
#Col.1 has whitespaces
#Col.2 has data entry error
#Col,31, 32, 33 data entry error
#Col.all NaNs
```

```python
# get the list of categorical descriptive features
categorical_cols = sw.columns[sw.dtypes==object].tolist()
categorical_cols

#Removing whitespaces from observations in categorical columns(if present)
for col in categorical_cols:
    print( 'Before transformation, unique values for ' + str(col))
    print(sw[col].unique())
    print('\nAfter transformation')
    sw[col] = sw[col].str.strip()
    print(sw[col].unique())
    print('')
```

```python
#Removing data entry errors in the data set

sw.replace(to_replace='Noo', value='No', inplace= True)
sw.replace(to_replace='Yess', value='Yes', inplace= True)
sw.replace(to_replace='no', value='No', inplace= True)
sw.replace(to_replace='yes', value='Yes', inplace= True)
sw.replace(to_replace='female', value='Female', inplace= True)
sw.replace(to_replace='male', value='Male',  inplace= True)
sw.replace(to_replace='F', value='Female',  inplace= True)
sw.replace(to_replace='M', value='Male', inplace= True)
```

```python
#Sanity check for 'Age' column
sw['Age'].unique()
```

```python
#Imputing extreme observation in 'Age' column with the mode
sw.loc[sw['Age'] == "500",'Age'] = sw['Age'].mode().values
```

```python
#Checking total NAs in each column
sw.isna().sum()

#Too many NAs per variable to omit/impute without skewing data set, hence they will be untouched
```

```python
#Mapping Boolean values to Columns 3-9 for ease of reading/accessing

true_false = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False,
}
for col in sw.columns[3:9]:
    sw[col] = sw[col].map(true_false)    
```

```python
#Mapping numeric grade scale for the Character Ranking columns 
#(Negative is unfavourable, 0 is neutral, Positive is favourable rank)

points = {
    "Very unfavorably": -2,
    "Somewhat unfavorably": -1,
    "Unfamiliar (N/A)": 0,
    "Neither favorably nor unfavorably (neutral)": 0,
    "Somewhat favorably": 1,
    "Very favorably": 2
}

for col in sw.columns[15:29]:
    sw[col] = sw[col].map(points)
    
sw.head()
```

```python
#Relationship between Age and Fan status of Star Wars
fanage_table = pd.crosstab(index=sw.iloc[:,2], 
                          columns=sw.iloc[:,34])

fanage_table
```

```python
#Plot - Relationship between Age and Fan status of Star Wars
fanage_table.plot(kind="bar", 
                 figsize=(8,8)
                 )
plt.ylabel('Frequency')
plt.title('Age-wise breakdown of Star Wars fans')
```

```python
#Relationship between Education and Ranking of Star Wars IV
edurank4_table = pd.crosstab(index=sw.iloc[:,13], 
                          columns=sw.iloc[:,36])

edurank4_table
```

```python
#Plot - Relationship between Education and Ranking of Star Wars IV
edurank4_table.plot(kind="bar", 
                 figsize=(8,8),
                 stacked= True)
plt.ylabel('Frequency')
plt.xlabel("Star Wars IV Ranking (1 highest - 6 lowest)")
plt.title("Ranking of Star Wars IV by Education level")
```

```python
#Relationship between Gender and Familiarity with the Expanded Universe
genderP_table = pd.crosstab(index=sw.iloc[:,33], 
                          columns=sw.iloc[:,30])

genderP_table
```

```python
#Plot - Relationship between Gender and Familiarity with the Expanded Universe
plot = genderP_table.plot.pie(y='Yes', figsize=(5, 5), autopct='%1.1f%%', startangle=60, pctdistance=0.85)
plt.title("Male and Female familiarity with the Expanded Universe", bbox={'facecolor':'0.8', 'pad':5})
plt.show()
```
